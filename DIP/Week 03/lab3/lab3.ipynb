```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
image = cv2.imread('ferrari_f1.avif')

# Check if loaded
if image is None:
    print("Error: ferrari_f1.avif not found or format not supported.")
    exit()

# Get height and width
(h, w) = image.shape[:2]

print("Image size:", h, "x", w)

# Function to display image
def show_image(img, title):
    plt.figure(figsize=(6,6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Show original image
show_image(image, "Original Image")
```

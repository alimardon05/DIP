```python
from PIL import Image
import numpy as np

img_path = "sample_data/ferrari_f1_2026.jpg"

try:
    # Open the image file using Pillow
    pil_img = Image.open(img_path)

    # Convert to grayscale ('L' mode) and resize to 256x256 pixels
    pil_img_processed = pil_img.convert('L').resize((256, 256))

    # Convert PIL Image to NumPy array
    image = np.array(pil_img_processed)

    print("Loaded image shape:", image.shape)
    print("Loaded image size (total elements):", image.size)

except FileNotFoundError:
    print(f"Error: The file '{img_path}' was not found.")
except Exception as e:
    print(f"An error occurred while loading or processing the image: {e}")
```

```python
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
plt.hist(image.flatten(), bins=256)
plt.title("Histogram of Pixel Intensities")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
smoothed = gaussian_filter(image, sigma=2)

plt.imshow(smoothed, cmap='gray')
plt.title("Gaussian Smoothed Image")
plt.axis('off')
plt.show()
```

```python
sx = ndimage.sobel(image, axis=0)
sy = ndimage.sobel(image, axis=1)
edges = np.hypot(sx, sy)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (Sobel)")
plt.axis('off')
plt.show()
```

```python
binary_image = image > 128

plt.imshow(binary_image, cmap='gray')
plt.title("Binary Image (Thresholding)")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(smoothed, cmap='gray')
plt.title("Smoothed")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(edges, cmap='gray')
plt.title("Edges")
plt.axis('off')

plt.show()
```
